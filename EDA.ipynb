```python
#chi-squared test is a statistical hypothesis test that is valid to perform when the test statistics 
#is chi-squared distributed under the null hypothesis,specifically pearson's chi-squared
#chi-square statistics is one way to show a relationship between categorical variaable.
#here we test correlation of categorical columns with target column

from scipy.stats import chi2_contingency
chi2_test=[]
for f in c_f:
    if chi2_contingency(pd.crosstab(df['selling_price'],df[f]))[1]<0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result=pd.DataFrame(data=[c_f,chi2_test]).T
result.columns=['Column','Hypothesis result']
result
```
